```python
df1_merge_df4_merge_df3_merge_df2.dropna(inplace=True)
```

```python
import numpy as np
```

```python
def percentil(product_category):
    category=df1_merge_df4_merge_df3_merge_df2[df1_merge_df4_merge_df3_merge_df2["Product_Category_Name"]==product_category]
    precios_cat=category["Price"].to_list()
    barato=np.percentile(precios_cat, 20)
    caro=np.percentile(precios_cat, 80)
    return barato, caro

categories = df2["Product_Category_Name"].unique()
percentiles = {c:percentil(c) for c in categories}
```

```python
def clasificar_precio(row):
    #print(rango)
    barato, caro = percentiles[row["Product_Category_Name"]]
    if row["Price"] <= barato:
        return 'barato'
    elif row['Price'] >= caro:
        return 'caro'
    else:
        return 'normal'

df1_merge_df4_merge_df3_merge_df2['nivel_precio'] = df1_merge_df4_merge_df3_merge_df2.apply(clasificar_precio, axis=1)
```

```python
import pandas as pd
from kmodes.kprototypes import KPrototypes
from sklearn.preprocessing import LabelEncoder

# Supongamos que ya tienes el DataFrame `df` filtrado por pas
features = ['Age', 'Gender', 'Product_Category_Name', 'Month', 'nivel_precio']
df_clustering = df_alemania[features].copy()

# Convertir columnas categricas a nmeros
cat_cols = ['Gender', 'Product_Category_Name', 'nivel_precio']
for col in cat_cols:
    df_clustering[col] = LabelEncoder().fit_transform(df_clustering[col])

# Convertir a numpy array
X = df_clustering.values

# ndices de columnas categricas
categorical_idx = [df_clustering.columns.get_loc(col) for col in cat_cols]

# Ajustar modelo K-Prototypes
kproto = KPrototypes(n_clusters=4, init='Huang', verbose=1, random_state=42)
clusters = kproto.fit_predict(X, categorical=categorical_idx)

# Aadir los clusters al DataFrame original
df_alemania['cluster'] = clusters
```

```python
# Ver la distribucin por cluster
df_alemania.groupby('cluster').agg({
    'Age':'mean',
    'Gender': lambda x: x.mode()[0],
    'Product_Category_Name': lambda x: x.mode()[0],
    'nivel_precio': lambda x: x.mode()[0],
    'Month': lambda x: x.mode()[0]
})
```

```python
df_alemania["cluster"].value_counts()
```

```python
import umap
import matplotlib.pyplot as plt

# UMAP trabaja solo con arrays numricos
reducer = umap.UMAP(random_state=42)
embedding = reducer.fit_transform(X)  # `X` es el array usado en K-Prototypes

# embedding es un array de (n_samples, 2)
```

```python
plt.figure(figsize=(10, 6))
scatter = plt.scatter(embedding[:, 0], embedding[:, 1], c=clusters, cmap='tab10', s=50, alpha=0.7)
plt.title('Clusters de usuarios proyectados en 2D con UMAP')
plt.xlabel('UMAP 1')
plt.ylabel('UMAP 2')
plt.colorbar(scatter, label='Cluster')
plt.grid(True)
plt.show()
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.cluster import KMeans
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

preprocessor = ColumnTransformer([
    ('num', StandardScaler(), ['Age',"Month"]),
    ('cat', OneHotEncoder(), ['Gender', 'Product_Category_Name',"nivel_precio"])
])
```

```python
pipeline = Pipeline([
    ('prep', preprocessor),
    ('kmeans', KMeans(n_clusters=10, random_state=42))
])

pipeline.fit(df_alemania)
df_alemania['cluster'] = pipeline.named_steps['kmeans'].labels_
```
